```python
import torch
from torchvision.models import resnet50
from PIL import Image
from torchvision import transforms as T
```

```python
model = resnet50(pretrained=True)
model.eval()
```

```python
img_path = './2.jpg'
```

```python
img = Image.open(img_path)
trans = T.Compose([T.Resize((224, 224)), T.ToTensor()])
timg = trans(img)
out = model(timg.unsqueeze(0))
pred = out[0].argmax()
```

```python
with open('imgnet.txt', 'r') as f:
    classes = [line.rstrip() for line in f]
    print(classes[pred])
```

```python
example_inputs = torch.rand(1, 3, 224, 224, requires_grad=True)
traced_foo = torch.jit.trace(model, example_inputs)
traced_foo.save('resnet50.jit')
```

```python
load_traced_model = torch.jit.load('resnet50.jit')
load_traced_model.eval()
```

```python
out1 = model(timg.unsqueeze(0))
out2 = load_traced_model(timg.unsqueeze(0))
```

```python
torch.linalg.norm(out1 - out2)
```
